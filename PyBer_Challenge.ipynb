# Pyber Challenge

### 4.3 Loading and Reading CSV files

```python
# Add Matplotlib inline magic command
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd

# File to Load (Remember to change these)
city_data_to_load = "Resources/city_data.csv"
ride_data_to_load = "Resources/ride_data.csv"

# Read the City and Ride Data
city_data_df = pd.read_csv(city_data_to_load)
ride_data_df = pd.read_csv(ride_data_to_load)
```

### Merge the DataFrames

```python
# Combine the data into a single dataset
pyber_data_df = pd.merge(ride_data_df, city_data_df, how="left", on=["city", "city"])

# Display the data table for preview
pyber_data_df.head()
```

## Deliverable 1: Get a Summary DataFrame 

```python
#  1. Get the total rides for each city type
rides_city_type_data_count = pyber_data_df.groupby(["type"]).count()["ride_id"]
rides_city_type_data_count.head()
```

```python
# 2. Get the total drivers for each city type
drivers_city_type_data_count = city_data_df.groupby(["type"]).sum()["driver_count"]
drivers_city_type_data_count.head()
```

```python
#  3. Get the total amount of fares for each city type
fares_city_type_data_count = pyber_data_df.groupby(["type"]).sum()["fare"]
fares_city_type_data_count.head()
```

```python
#  4. Get the average fare per ride for each city type. 
avg_fare_ride_city_type = fares_city_type_data_count/rides_city_type_data_count
avg_fare_ride_city_type.head()
```

```python
# 5. Get the average fare per driver for each city type. 
avg_fare_driver_city_type = fares_city_type_data_count/drivers_city_type_data_count
avg_fare_driver_city_type.head()
```

```python
#  6. Create a PyBer summary DataFrame. 
pyber_summary_df = pd.DataFrame({
    "Total Rides" : rides_city_type_data_count,
    "Total Drivers" : drivers_city_type_data_count,
    "Total Fares" : fares_city_type_data_count,
    "Average Fare per Ride" : avg_fare_ride_city_type,
    "Average Fare per Driver" : avg_fare_driver_city_type
                             })
pyber_summary_df
```

```python
#  7. Cleaning up the DataFrame. Delete the index name
pyber_summary_df.index.name = None
```

```python
#  8. Format the columns.
pyber_summary_df.loc[:, "Total Rides"] = pyber_summary_df["Total Rides"].map('{:,}'.format)
pyber_summary_df.loc[:, "Total Drivers"] = pyber_summary_df["Total Drivers"].map('{:,}'.format)
pd.options.display.float_format = '${:,.2f}'.format
pyber_summary_df
```

## Deliverable 2.  Create a multiple line plot that shows the total weekly of the fares for each type of city.

```python
# 1. Read the merged DataFrame
# Read the City and Ride Data
city_data_df = pd.read_csv(city_data_to_load)
ride_data_df = pd.read_csv(ride_data_to_load)
# Combine the data into a single dataset
new_pyber_data_df = pd.merge(ride_data_df, city_data_df, how="left", on=["city", "city"])
new_pyber_data_df
```

```python
# 2. Using groupby() to create a new DataFrame showing the sum of the fares 
#  for each date where the indices are the city type and date.
sum_fares_date_city_date = pd.DataFrame({"fare": new_pyber_data_df.groupby(["type","date"]).sum()["fare"]})
sum_fares_date_city_date
```

```python
# 3. Reset the index on the DataFrame you created in #1. This is needed to use the 'pivot()' function.
sum_fares_date_city_date_reset = sum_fares_date_city_date.reset_index()
sum_fares_date_city_date_reset
```

```python
# 4. Create a pivot table with the 'date' as the index, the columns ='type', and values='fare' 
# to get the total fares for each type of city by the date. 
pivot_sum_fares_date_city_date_reset = pd.pivot_table(sum_fares_date_city_date_reset, values=["fare"], index=["date"],
                    columns=['type'] )
pivot_sum_fares_date_city_date_reset.head()
```

```python
# 5. Create a new DataFrame from the pivot table DataFrame using loc on the given dates, '2019-01-01':'2019-04-28'.

new_pivot_sum_fares_date_city_date_reset = pivot_sum_fares_date_city_date_reset.loc["2019-01-01":"2019-04-28"]
new_pivot_sum_fares_date_city_date_reset
```

```python
# 6. Set the "date" index to datetime datatype. This is necessary to use the resample() method in Step 8.
new_pivot_sum_fares_date_city_date_reset.index = pd.to_datetime(new_pivot_sum_fares_date_city_date_reset.index)
```

```python
# 7. Check that the datatype for the index is datetime using 
new_pivot_sum_fares_date_city_date_reset.info()
```

```python
# 8. Create a new DataFrame using the "resample()" function by week 'W' and get the sum of the fares for each week.
resampled_new_pivot_sum_fares_date_city_date_reset = new_pivot_sum_fares_date_city_date_reset.resample('W').sum()
resampled_new_pivot_sum_fares_date_city_date_reset
```

```python
# 8. Using the object-oriented interface method, plot the resample DataFrame using the df.plot() function. 

# Import the style from Matplotlib.
from matplotlib import style
# Use the graph style fivethirtyeight.
style.use('fivethirtyeight')

# Plot Multiple line graph using Five Thirty Eight Style
fig, ax= plt.subplots(figsize = (25,8))

# Plotting Fares for Rural, Suburban & Urban City Type
ax.plot(resampled_new_pivot_sum_fares_date_city_date_reset[('fare', 'Rural')],label = "Rural")
ax.plot(resampled_new_pivot_sum_fares_date_city_date_reset[('fare', 'Suburban')],label = "Suburban")
ax.plot(resampled_new_pivot_sum_fares_date_city_date_reset[('fare', 'Urban')],label = "Urban")

# Add the title, y-axis label and grid.
ax.set_title("Total Fare by City Type", size = 20)
ax.set_ylabel("Fare($USD)", size = 15)
ax.legend(loc = "center", title = "type")
plt.xticks(['2019-01-06','2019-02-03','2019-03-03','2019-04-07'], ["Jan 2019"," Feb", "Mar", "Apr"])

# Save the figure.
plt.savefig("analysis/PyBer_fare_summary.png")
plt.show()
```
